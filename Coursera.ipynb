## Building the dataset
Our dataset consists of selecting various streets within about a 10 minute walk of O'Connel street and then retrieving all venues under the "food" categories returned by Foursquare for each address.

```python
address_list = ["O'Connell street", "Grafton Street" , "Cook Street", "Henry Street", "Parnell Street", "College Green", "Temple Bar", "Fleet street", "Talbot Street", "Capel Street","Essex Street", "North Earl Street", "wicklow Street", "Molesworth Street", "Nassau Street", "Fenian Street", "Hanover Street", "Great Strand Street", "Inns Quay", "Amiens Street"]
CATEGORY = '4d4b7105d754a06374d81259'
LIMIT = 50
radius = 500

frames = []

for address in address_list:
    geolocator = Nominatim(user_agent="test_app")
    location = geolocator.geocode(address + ', Dublin,Ireland')
    latitude = location.latitude
    longitude = location.longitude

    url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&categoryId={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, CATEGORY, radius, LIMIT)
    
    results = requests.get(url).json()
    venues = results['response']['venues']
    dataframe = json_normalize(venues)
    frames.append(dataframe)
    
result = pd.concat(frames)
result  
```

## Data Cleanup

Let's change that JSON string to the more readable name attribute within the string.
While we're at it we should rename columns, check for duplicates and remove them and reset our index which is currently all over the place.

```python
result["categories"] = result["categories"].agg(lambda x: x[0].get('name'))  #  Get the catagory name field.
```

```python
result = result[['location.lng','location.lat','name','id', 'categories']]
result.columns = ["longitude", "latitude", "name", "id", "categories"]
result.drop_duplicates(subset=['id']) 
result.reset_index(inplace=True)
result = result.drop('index', 1)



result
```
